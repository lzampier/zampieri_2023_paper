## Importing Python Packages 

```python
import os
import pickle
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from scipy.stats import binned_statistic
```

## Useful parameters 

```python
# Densities (kg m-3).
rhoi = 917
rhow = 1026
rhos = 330

# Thermal Conductivity Values (W m-1 K-1).
k_ice = 2.1   # Sea ice conductivity
k_snow_3  = 0.3   # Snow conductivity (0.3 W m-1 K-1)
k_snow_1  = 0.1   # Snow conductivity (0.1 W m-1 K-1)

# Ocean Freezing Temperature in K (T_f = -1.8 K).
T_oce = 273.15 - 1.8

# Transect Parameters.
step = 1.1

# ITD discretization (m).
model_itd_bins = np.array([0.0, 0.6, 1.4, 2.4, 3.6, 100.0])

# Save panels. Do you want to save the panel file?
save_file=False

# Dates corresponding to each transect file.
file_array =         [['20191226T120647', '20200107T075304', '20200115T122403', '20200116T115614',
                       '20200119T112023', '20200126T131018', '20200320T122848', '20200424T115507',
                       '20200424T124835', '20200424T130256'],
                      ['20191225T140809', '20200107T092304', '20200116T063418', '20200116T063418',
                       '20200121T103600', '20200128T063509', '20200321T081331', '20200423T074432',
                       '20200423T074432', '20200423T074432'],
                      ['20191225T154011', '20200107T105236', '20200116T075650', '20200116T075650', 
                       '20200121T122810', '20200128T080711', '20200321T100825', '20200423T093134',
                       '20200423T093134', '20200423T093134']]
```

## Panel 1 processing

```python
# We choose the transect in position one of the previous list corresponding to '20200107T075304' by setting ind_transect.
ind_transect = 3
transect_string = file_array[0][ind_transect]
plot_date = '{}.{}.{}'.format(transect_string[0:4],transect_string[4:6],transect_string[6:8])
datetime_1 = file_array[1][ind_transect]
datetime_2 = file_array[2][ind_transect]
transect_file = './data/transects/transect_{}.pkl'.format(transect_string)

if not os.path.isfile(transect_file):
    print('ERROR: Transect file does not exhist for {}!'.format(transect_string))
else:
    print('OK: File available for model computations. Reading the transect variables.')    
    with open(transect_file,'rb') as f:
        transect = pickle.load(f)
    # Reading transect variables.
    fr_transect = transect[:,0]
    ts_transect = transect[:,1]
    hi_transect = transect[:,2]
    hs_transect = transect[:,3]
    # Cutoff for thin snow. We want to maintain a minimum snow thickness of for numerical simplicity 2 cm.
    hs_transect[hs_transect < 0.02] = 0.02
    # Snow freeboard from thickness measurements â€“ Constant density assumption.
    sf_transect = ((rhow - rhoi)/rhow)*hi_transect + ((rhow - rhos)/rhow)*hs_transect
    ps_transect = np.arange(0, transect[:,:].shape[0])*step  
    # Model discretization of the transect.
    vicen = binned_statistic(hi_transect, hi_transect, statistic='mean', bins=model_itd_bins).statistic
    vsnon = binned_statistic(hi_transect, hs_transect, statistic='mean', bins=model_itd_bins).statistic
    unit = hi_transect.copy()
    unit[:] = 1
    aicen = binned_statistic(hi_transect, unit, statistic='sum', bins=model_itd_bins).statistic

hi_transect_cesm = hi_transect.copy()
hs_transect_cesm = hi_transect.copy()
hi_transect_cesm_1d = hi_transect.copy()
hs_transect_cesm_1d = hi_transect.copy()
for i in range(0,len(hi_transect)):
    hi_transect_cesm_1d[i] = np.nanmean(vicen)
    hs_transect_cesm_1d[i] = np.nanmean(vsnon)    
    if i < np.sum(aicen[0:1]):
        hi_transect_cesm[i] = vicen[0]
        hs_transect_cesm[i] = vsnon[0]
    elif i < np.sum(aicen[0:2]):
        hi_transect_cesm[i] = vicen[1]
        hs_transect_cesm[i] = vsnon[1]
    elif i < np.sum(aicen[0:3]):
        hi_transect_cesm[i] = vicen[2]
        hs_transect_cesm[i] = vsnon[2]
    elif i < np.sum(aicen[0:4]):
        hi_transect_cesm[i] = vicen[3]
        hs_transect_cesm[i] = vsnon[3]
    elif i < np.sum(aicen[0:5]):
        hi_transect_cesm[i] = vicen[4]
        hs_transect_cesm[i] = vsnon[4]

# Read transect temperature profile 
data_folder = './data/diagnostics/{}_data/'.format(transect_string)

with open('{}solution_2d_{}_emiss1.pkl'.format(data_folder, 'k3'),'rb') as f:
    [T_half_ice, T_half_snow, T_top, T_bottom, T_interface, T_semibot_ice] = pickle.load(f)
    
T_half_ice[T_half_ice==0] = np.nan
T_half_snow[T_half_snow==0] = np.nan
T_top[T_top==0] = np.nan
T_bottom[T_bottom==0] = np.nan
T_interface[T_interface==0] = np.nan

with open('{}solution_1d_{}.pkl'.format(data_folder, 'k3'),'rb') as f:
    [T_half_ice_1d, T_half_snow_1d, T_top_1d, T_bottom_1d, T_interface_1d, T_semibot_ice_1d] = pickle.load(f)
    
T_half_ice_1d[T_half_ice_1d==0] = np.nan
T_half_snow_1d[T_half_snow_1d==0] = np.nan
T_top_1d[T_top_1d==0] = np.nan
T_bottom_1d[T_bottom_1d==0] = np.nan
T_interface_1d[T_interface_1d==0] = np.nan
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(18,13))
axs[0,0].plot(ps_transect, -hi_transect, color='tab:blue', label='â€“ Sea Ice Thickness')
axs[0,0].fill_between(ps_transect, -hi_transect, color='tab:blue',alpha=0.4)
axs[0,0].fill_between(ps_transect, -hi_transect, -100-hi_transect, color='tab:blue',alpha=0.1)
axs[0,0].plot(ps_transect, hs_transect*10, color='tab:cyan', label='Snow Thickness x 10')
axs[0,0].fill_between(ps_transect, hs_transect*10, color='tab:cyan',alpha=0.4)
axs[0,0].fill_between(ps_transect, hs_transect*10, 100+hs_transect*10, color='black',alpha=0.5)
axs[0,0].plot(ps_transect, np.zeros(ps_transect.shape), color='tab:gray', label='Snow-Ice Interface')
axs[0,0].set_ylabel('Height (m)')
axs[0,0].set_xlabel('Transect Length (m)')
axs[0,0].set_title('(a)  Snow and Sea Ice Thickness from Transect\n'+plot_date)
axs[0,0].set_ylim((-10,8))
axs[0,0].set_xlim((ps_transect[0],ps_transect[-1]))
axs[0,0].legend()    

axs[1,0].plot(ps_transect, T_interface_1d, 'tab:gray')
axs[1,0].plot(ps_transect, T_top_1d, 'tab:cyan')
axs[1,0].plot(ps_transect, T_bottom_1d, 'tab:blue')
axs[1,0].plot(ps_transect, ts_transect, 'k')
axs[1,0].set_ylabel('Temperature (K)')
axs[1,0].set_xlabel('Transect Length (m)')
axs[1,0].set_title('(c)  1D Vertical Thermodynamic Solution â€“ '+plot_date)
axs[1,0].set_xlim((ps_transect[0],ps_transect[-1]))

axs[0,1].plot(ps_transect/ps_transect.shape[0]/1.1, -hi_transect_cesm, color='tab:blue')
axs[0,1].fill_between(ps_transect/ps_transect.shape[0]/1.1, -hi_transect_cesm, color='tab:blue',alpha=0.4)
axs[0,1].fill_between(ps_transect/ps_transect.shape[0]/1.1, -hi_transect_cesm, -100-hi_transect_cesm, color='tab:blue',alpha=0.1)
axs[0,1].plot(ps_transect/ps_transect.shape[0]/1.1, hs_transect_cesm*10, color='tab:cyan')
axs[0,1].fill_between(ps_transect/ps_transect.shape[0]/1.1, hs_transect_cesm*10, color='tab:cyan',alpha=0.4)
axs[0,1].fill_between(ps_transect/ps_transect.shape[0]/1.1, hs_transect_cesm*10, 100+hs_transect*10, color='black',alpha=0.5)
axs[0,1].plot(ps_transect/ps_transect.shape[0]/1.1, np.zeros(ps_transect.shape), color='tab:gray')
axs[0,1].set_ylabel('Height (m)')
axs[0,1].set_xlabel('Fractional Coverage of Transect Length')
axs[0,1].set_title('(b)  ITD Discretized Snow and Sea Ice Thickness from Transect\n'+plot_date)
axs[0,1].set_ylim((-10,8))
axs[0,1].set_xlim((ps_transect[0]/1.1/ps_transect.shape[0],ps_transect[-1]/1.1/ps_transect.shape[0]))

axs[1,1].plot(ps_transect, T_interface, 'tab:gray')
axs[1,1].plot(ps_transect, T_top, 'tab:cyan', label='T$_{SA}$')
axs[1,1].plot(ps_transect, T_bottom, 'tab:blue', label='T$_{OI}$')
axs[1,1].plot(ps_transect, ts_transect, 'k', label='T$_{SA~Obs}$')
axs[1,1].set_ylabel('Temperature (K)')
axs[1,1].set_xlabel('Transect Length (m)')
axs[1,1].legend(loc = 'upper right')
axs[1,1].set_title('(d)  2D Thermodynamic Solution â€“ '+plot_date)
axs[1,1].set_xlim((ps_transect[0],ps_transect[-1]))

for int_itd in range(0,5):
    axs[0,1].scatter(10, -100, s = 0.1, marker='o', color='white', alpha=0, 
                label='ITD Class '+str(int_itd+1)+'   '+ str(model_itd_bins[int_itd]) +'m - '+ str(model_itd_bins[int_itd+1]) +'m')
axs[0,1].legend(loc='lower left')

if save_file:
    plt.savefig('panel_1.pdf', bbox_inches='tight')
plt.show()
```

## Panel 2 â€“ Part a and b

```python
itd_average = []
itd_model = []
itd_model_1d = []
itd_tgradient = []
itd_aicen = []
sizing_factor = []

for ind_transect in range(0,len(file_array[0])):
    transect_string = file_array[0][ind_transect]
    datetime_1 = file_array[1][ind_transect]
    datetime_2 = file_array[2][ind_transect]
    transect_file = './transects/transect_{}.pkl'.format(transect_string)

    if not os.path.isfile(transect_file):
        nothing = 0
        #print('ERROR: Transect file does not exhist for {}!'.format(transect_string))
    else:
        #print('OK: File available for model computations. Reading the transect variables.')    
        with open(transect_file,'rb') as f:
            transect = pickle.load(f)

        fr_transect = transect[:,0]
        ts_transect = transect[:,1]
        hi_transect = transect[:,2]
        hs_transect = transect[:,3]
        # Cutoff for thin snow. We want to maintain a minimum snow thickness
        hs_transect[hs_transect < 0.02] = 0.02
        # Snow freeboard from thickness measurements â€“ Constant density assumption
        sf_transect = ((rhow - rhoi)/rhow)*hi_transect + ((rhow - rhos)/rhow)*hs_transect
        ps_transect = np.arange(0, transect[:,:].shape[0])*step    
        vicen = binned_statistic(hi_transect, hi_transect, statistic='mean', bins=model_itd_bins).statistic
        vsnon = binned_statistic(hi_transect, hs_transect, statistic='mean', bins=model_itd_bins).statistic
        thericen = vicen + (vsnon*k_ice/k_snow_3)

    data_folder = './diagnostics/{}_data/'.format(file_array[0][ind_transect])
    with open('{}solution_1d_k3.pkl'.format(data_folder),'rb') as f:
        [T_half_ice_1d, T_half_snow_1d, T_top_1d, T_bottom_1d, T_interface_1d, T_semibot_ice_1d] = pickle.load(f)

    with open('{}solution_2d_k3_emiss1.pkl'.format(data_folder),'rb') as f:
        [T_half_ice, T_half_snow, T_top, T_bottom, T_interface, T_semibot_ice] = pickle.load(f)

    if os.path.isfile('./diagnostics/{}_data/itd_solution_1d_k3.pkl'.format(transect_string)):
        with open('./diagnostics/{}_data/itd_solution_1d_k3.pkl'.format(transect_string),'rb') as f:
            [itd_T_half_ice, itd_T_half_snow, itd_T_top, itd_T_bottom, itd_T_interface, itd_T_semibot_ice] = pickle.load(f)

    bot_temp = - 1.8 + 273.15
    itd_gradient = (itd_T_interface - bot_temp)/vicen
    itd_conductive = k_ice * ((itd_T_semibot_ice[:] - itd_T_bottom[:])/0.1)

    bot_temp = - 1.8 + 273.15
    t3_grad_snow = (T_top[:] - T_interface[:])/hs_transect
    t3_grad_ice = (T_interface[:] - bot_temp)/hi_transect

    
    bot_temp = - 1.8 + 273.15
    t3_grad_snow_1d = (T_top_1d[:] - T_interface_1d[:])/hs_transect
    t3_grad_ice_1d = (T_interface_1d[:] - bot_temp)/hi_transect

    y3_mod = hi_transect + hs_transect*k_ice/k_snow_3

    conductive = k_ice * ((T_semibot_ice[:] - T_bottom[:])/0.1)
    conductive_1d = k_ice * ((T_semibot_ice_1d[:] - T_bottom_1d[:])/0.1)
    cicen = binned_statistic(hi_transect, conductive, statistic='mean', bins=model_itd_bins).statistic
    cicen_1d = binned_statistic(hi_transect, conductive_1d, statistic='mean', bins=model_itd_bins).statistic
    unit = conductive.copy()
    unit[:] = 1
    aicen = binned_statistic(hi_transect, unit, statistic='sum', bins=model_itd_bins).statistic/unit.shape[0]
    
    if hi_transect.shape[0]>0:
        itd_model.append(cicen)
        itd_model_1d.append(cicen_1d)
        itd_average.append(itd_conductive)
        itd_tgradient.append(itd_gradient)
        itd_aicen.append(aicen*conductive.shape[0])

fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(5,5))

markerlist = ['o','*','X','^','P']
colorlist = ['#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f',
             '#ff7f00','#cab2d6','#6a3d9a','#ffff99']

for ind_transect in range(0,len(itd_tgradient)):
    for int_itd in range(0,len(cicen)):
        axs.scatter(itd_average[ind_transect][int_itd], itd_model_1d[ind_transect][int_itd]-itd_average[ind_transect][int_itd],
                    s=itd_aicen[ind_transect][int_itd]/1.7, marker=markerlist[int_itd], color=colorlist[ind_transect],
                    edgecolor='black', linewidth=0.3, zorder=5)

for ind_transect in range(0,len(itd_tgradient)):
    axs.scatter(10, 0, s = 80, marker='o', color=colorlist[ind_transect], label=file_array[0][ind_transect][0:8])

axs.plot([0,0],[-1000,100],color='tab:grey')  
axs.plot([-100,100],[0,0],color='tab:grey')  

axs.set_xlim((-40,0))
axs.set_ylim((-8,4))
axs.legend(loc='lower left', ncol=2)

axs.set_ylabel('Conductive Heat Flux Difference (W m$^{-2}$)')
axs.set_xlabel('ITD Conductive Heat Flux (W m$^{-2}$)')
axs.set_title('(a)  1D â€“ 1D$_{ITD}$ Thermodynamic Solutions')

if save_file:
    plt.savefig('panel_2_1.pdf', bbox_inches='tight', dpi=200)
plt.show()

fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(5,5))

markerlist = ['o','*','X','^','P']
colorlist = ['#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f',
             '#ff7f00','#cab2d6','#6a3d9a','#ffff99']

for ind_transect in range(0,len(itd_tgradient)):
    for int_itd in range(0,len(cicen)):
        axs.scatter(itd_average[ind_transect][int_itd], itd_model[ind_transect][int_itd]-itd_average[ind_transect][int_itd],
                    s=itd_aicen[ind_transect][int_itd]/1.7, marker=markerlist[int_itd], color=colorlist[ind_transect],
                    edgecolor='black', linewidth=0.3, zorder=5)

axs.plot([0,0],[-1000,100],color='tab:grey')  
axs.plot([-100,100],[0,0],color='tab:grey')  

axs.set_xlim((-40,0))
axs.set_ylim((-8,4))

axs.set_ylabel('Conductive Heat Flux Difference (W m$^{-2}$)')
axs.set_xlabel('ITD Conductive Heat Flux (W m$^{-2}$)')
axs.set_title('(b)  2D â€“ 1D$_{ITD}$ Thermodynamic Solutions')

if save_file:
    plt.savefig('panel_2_2.pdf', bbox_inches='tight', dpi=300)
plt.show()
```

## Panel 3 processing

```python
# We choose the transect in position one of the previous list corresponding to '20200107T075304' by setting ind_transect.
ind_transect = 1
transect_string = file_array[0][ind_transect]
plot_date = '{}.{}.{}'.format(transect_string[0:4],transect_string[4:6],transect_string[6:8])
datetime_1 = file_array[1][ind_transect]
datetime_2 = file_array[2][ind_transect]
transect_file = './transects/transect_{}.pkl'.format(transect_string)

if not os.path.isfile(transect_file):
    print('ERROR: Transect file does not exhist for {}!'.format(transect_string))
else:
    print('OK: File available for model computations. Reading the transect variables.')    
    with open(transect_file,'rb') as f:
        transect = pickle.load(f)
    # Reading transect variables.
    fr_transect = transect[:,0]
    ts_transect = transect[:,1]
    hi_transect = transect[:,2]
    hs_transect = transect[:,3]
    # Cutoff for thin snow. We want to maintain a minimum snow thickness of for numerical simplicity 2 cm.
    hs_transect[hs_transect < 0.02] = 0.02
    # Snow freeboard from thickness measurements â€“ Constant density assumption.
    sf_transect = ((rhow - rhoi)/rhow)*hi_transect + ((rhow - rhos)/rhow)*hs_transect
    ps_transect = np.arange(0, transect[:,:].shape[0])*step  
    # Model discretization of the transect.
    vicen = binned_statistic(hi_transect, hi_transect, statistic='mean', bins=model_itd_bins).statistic
    vsnon = binned_statistic(hi_transect, hs_transect, statistic='mean', bins=model_itd_bins).statistic
    unit = hi_transect.copy()
    unit[:] = 1
    aicen = binned_statistic(hi_transect, unit, statistic='sum', bins=model_itd_bins).statistic

# Reading SIMBA data and computing the temperature gradients for the cases k1 and k3

with open('SIMB-temperatures_helicopter-times_20230412.pkl','rb') as f:
    SIMB_file = pickle.load(f)

grad_snow = []
grad_ice  = []
snow = []
ice  = []

for key in SIMB_file[datetime_1]:
    int_temp = SIMB_file[datetime_1][key]['ice_snow_interface_temp'] + 273.15
    top_temp = SIMB_file[datetime_1][key]['snow_top_temp'] + 273.15
    bot_temp = T_oce
    ice.append(SIMB_file[datetime_1][key]['snow_depth']/100)
    snow.append(SIMB_file[datetime_1][key]['ice_thickness']/100)
    grad_snow.append((top_temp - int_temp) / snow[-1])
    grad_ice.append((int_temp - bot_temp) / ice[-1])

y3_obs = np.array(ice) + np.array(snow)*k_ice/k_snow_3
y1_obs = np.array(ice) + np.array(snow)*k_ice/k_snow_1

# Read transect temperature profile 
data_folder = './diagnostics/{}_data/'.format(transect_string)

# Read transect temperature data
with open('{}solution_2d_{}_emiss1.pkl'.format(data_folder, 'k3'),'rb') as f:
    [T_half_ice, T_half_snow, T_top, T_bottom, T_interface, T_semibot_ice] = pickle.load(f)

# Remove possible NaNs
T_half_ice[T_half_ice==0] = np.nan
T_half_snow[T_half_snow==0] = np.nan
T_top[T_top==0] = np.nan
T_bottom[T_bottom==0] = np.nan
T_interface[T_interface==0] = np.nan

# Read transect temperature data
with open('{}solution_2d_{}_emiss1.pkl'.format(data_folder, 'k1'),'rb') as f:
    [T_half_ice_1, T_half_snow_1, T_top_1, T_bottom_1, T_interface_1, T_semibot_ice_1] = pickle.load(f)

# Remove possible NaNs
T_half_ice_1[T_half_ice_1==0] = np.nan
T_half_snow_1[T_half_snow_1==0] = np.nan
T_top_1[T_top_1==0] = np.nan
T_bottom_1[T_bottom_1==0] = np.nan
T_interface_1[T_interface_1==0] = np.nan

# Compute the gradients within the ice at half of the ice thcikness
gradient_k1 = -(T_bottom_1 - T_interface_1) / hi_transect
gradient_k3 = -(T_bottom - T_interface) / hi_transect

# Compute the effective thermodynamic thickness for the transect
y3_mod = hi_transect + hs_transect*k_ice/k_snow_3
y1_mod = hi_transect + hs_transect*k_ice/k_snow_1

# Compute the conductive heat flux within the ice
conductive = k_ice * ((T_semibot_ice[:] - T_bottom[:])/0.1)
conductive_1d = k_ice * ((T_semibot_ice_1[:] - T_bottom[:])/0.1)
```
